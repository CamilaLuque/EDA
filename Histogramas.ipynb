```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv('/distancias_10.txt', sep='/n', header=None)
dim2 = df2.T.squeeze()
dim2.plot.hist(grid=True, bins=20, rwidth=0.9, color='#607c8e')
plt.title('Histograma de las distancias entre puntos con dimensi√≥n 10')
plt.xlabel('Distancia')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.xlim(0,2)
plt.show()
```
